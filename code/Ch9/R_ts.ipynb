```r
(co2.linear.model <- lm(co2 ~ time(co2)))
```

```r
plot(co2, main="Atmospheric CO@ Concentration with Fitted Line")
abline(co2.linear.model)
```

```r
co2.residuals <- resid(co2.linear.model)
hist(co2.residuals, main="Historgram of Residuals")
```

```r
qqnorm(co2.residuals)
qqline(co2.residuals)
#it shows below that we have some deviation from normality
```

```r
#doing a time plot of our residuals 
plot(co2.residuals ~ time(co2))
```

```r
# Zoom into our residuals
plot(co2.residuals ~ time(co2), xlim = c(1960, 1963), main="Zoomed Residuals")
```

```r
x <- NULL
x[1] = 0
for (i in 2:1000) {
    x[i] <- x[i-1]+rnorm(1)
}

random_walk <- ts(x) #not stationary time series

plot(random_walk, main='A random walk', ylab='', xlab='Days', col='blue', lwd=2)
```

```r
acf(random_walk)
```

```r
plot(diff(random_walk)) #should give me purely random process - white noise
```

```r
milk <- read.csv('../../datasets/Ch9/milk_production.csv', header = TRUE, stringsAsFactors = FALSE)
head(milk)
```

```r
milk_tbl <- as.tibble(milk)
```

```r
milk_ts <- as.ts(milk_tbl$production, frequency = 12)
```

```r
milk.ts <- as.ts(data = milk$production, frequency = 12, start=c(1962,1), end=c(1975,12))
head(milk.ts)
```

```r
plot(milk_ts)
```

```r
gg_tsdisplay(difference(milk_ts, 12),
               plot_type='partial', lag=36) +
  labs(title="Seasonally differenced", y="")
```

```r
leisure <- us_employment %>%
  filter(Title == "Leisure and Hospitality",
         year(Month) > 2000) %>%
  mutate(Employed = Employed/1000) %>%
  select(Month, Employed)
autoplot(leisure, Employed) +
  labs(title = "US employment: leisure and hospitality",
       y="Number of people (millions)")
```

```r
leisure %>%
  gg_tsdisplay(difference(Employed, 12),
               plot_type='partial', lag=36) +
  labs(title="Seasonally differenced", y="")
```

```r
leisure %>%
  gg_tsdisplay(difference(Employed, 12) %>% difference(),
               plot_type='partial', lag=36) +
  labs(title = "Double differenced", y="")
```

```r
fit <- leisure %>%
  model(
    arima012011 = ARIMA(Employed ~ pdq(0,1,2) + PDQ(0,1,1)),
    arima210011 = ARIMA(Employed ~ pdq(2,1,0) + PDQ(0,1,1)),
    auto = ARIMA(Employed, stepwise = FALSE, approx = FALSE)
  )
fit %>% pivot_longer(everything(), names_to = "Model name",
                     values_to = "Orders")
#> # A mable: 3 x 2
#> # Key:     Model name [3]
#>   `Model name`                    Orders
#>   <chr>                          <model>
#> 1 arima012011  <ARIMA(0,1,2)(0,1,1)[12]>
#> 2 arima210011  <ARIMA(2,1,0)(0,1,1)[12]>
#> 3 auto         <ARIMA(2,1,0)(1,1,1)[12]>
glance(fit) %>% arrange(AICc) %>% select(.model:BIC)
#> # A tibble: 3  6
#>   .model       sigma2 log_lik   AIC  AICc   BIC
#>   <chr>         <dbl>   <dbl> <dbl> <dbl> <dbl>
#> 1 auto        0.00142    395. -780. -780. -763.
#> 2 arima210011 0.00145    392. -776. -776. -763.
#> 3 arima012011 0.00146    391. -775. -775. -761.
```

```r
fit %>% select(auto) %>% gg_tsresiduals(lag=36)
```
