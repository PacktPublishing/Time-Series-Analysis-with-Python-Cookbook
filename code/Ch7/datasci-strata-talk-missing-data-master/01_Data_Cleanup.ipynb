## Handling the Gap

The gap in our data shows up as adjacent rows with a long duration between their timestamps.  Add in rows, sampled every 15 minutes, with NaN to make it easier to show the different methods in future.

```python
df = pd.DataFrame(glucose_ser)
df
```

```python
gap_sizes = {}
for pos, val in enumerate(list(df.index)):
    if pos > 0:
        gap_sizes[pos-1] = val - list(df.index)[pos-1]
gap_sizes
```

```python
GAP_LEFT_INDEX = 59
GAP_RIGHT_INDEX = 60
GAP_LEFT_MINUTES = list(df.index)[GAP_LEFT_INDEX]
GAP_RIGHT_MINUTES = list(df.index)[GAP_RIGHT_INDEX]
GAP_LEFT_INDEX, GAP_RIGHT_INDEX, GAP_LEFT_MINUTES, GAP_RIGHT_MINUTES
```

```python
missing_indices = np.floor((GAP_RIGHT_MINUTES-GAP_LEFT_MINUTES)/15).astype(int)
missing_minutes = [GAP_LEFT_MINUTES + i*15 for i in range(missing_indices+1)]
missing_minutes
```

```python
new_index_labels = sorted(list(df.index)+missing_minutes)
new_index_labels
```

```python
df = df.reindex(new_index_labels)
df
```
